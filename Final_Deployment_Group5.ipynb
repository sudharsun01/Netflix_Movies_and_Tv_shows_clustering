```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Select relevant columns
netflix_data = netflix[['type', 'title', 'director', 'cast', 'country', 'rating', 'listed_in', 'release_year', 'day', 'month', 'year']].copy()

# Convert all string columns to lowercase
netflix_data = netflix_data.applymap(lambda x: x.lower() if isinstance(x, str) else x)

# Encode categorical columns
label_encoder = LabelEncoder()
for column in ['type', 'rating', 'listed_in']:
    netflix_data[column] = label_encoder.fit_transform(netflix_data[column])

# Convert encoded values back to strings
for column in ['type', 'rating', 'listed_in']:
    netflix_data[column] = label_encoder.inverse_transform(netflix_data[column])

# Create a text representation of the combined categorical columns
netflix_data['combined'] = netflix_data[['type', 'title', 'director', 'cast', 'country', 'rating', 'listed_in']].agg(' '.join, axis=1)

# Convert text data to TF-IDF features
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(netflix_data['combined'])

# Apply k-means clustering
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
netflix_data['cluster'] = kmeans.fit_predict(tfidf_matrix)
```

```python
# Reduce dimensionality for visualization
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(tfidf_matrix.toarray())
netflix_data['pca1'] = reduced_features[:, 0]
netflix_data['pca2'] = reduced_features[:, 1]

# Plot clusters in 2D
plt.scatter(netflix_data['pca1'], netflix_data['pca2'], c=netflix_data['cluster'], cmap='viridis', alpha=0.5)
plt.title('K-Means Clustering of Netflix Data')
#plt.xlabel('PCA1')
#plt.ylabel('PCA2')
#plt.show()
```

```python
import pickle

# Save the k-means model
with open('kmeans_model.pkl', 'wb') as model_file:
    pickle.dump(kmeans, model_file)

# Save the preprocessed data
with open('preprocessed_data.pkl', 'wb') as data_file:
    pickle.dump(netflix_data, data_file)
```
